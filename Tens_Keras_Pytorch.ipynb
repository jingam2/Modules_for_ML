```python
'''
딥러닝 개요
딥러닝은 인공지능(AI)과 머신러닝(ML)의 한 분야로, 인간의 뇌 구조와 유사한 인공신경망을 사용해 데이터를 학습합니다.
딥러닝의 주요 목표는 주어진 데이터를 기반으로 패턴을 인식하고, 이를 통해 예측이나 분류 등의 작업을 수행하는 것입니다.

인공신경망 기본 개념
인공신경망(ANN)은 뇌의 뉴런 구조를 모방한 모델입니다.
ANN은 여러 개의 노드(뉴런)로 구성되며, 각 노드는 입력 값을 받아 특정 가중치를 곱하고 활성화 함수를 통해 출력 값을 계산합니다.
여러 층(layer)을 쌓아 더 복잡한 문제를 해결할 수 있습니다.
'''
```

```python
'''
퍼셉트론과 다층 퍼셉트론

퍼셉트론은 인공신경망의 가장 기본적인 단위입니다.
입력 값에 가중치를 곱하고, 이를 모두 더한 다음 활성화 함수를 통해 결과를 출력합니다.
다층 퍼셉트론(MLP)은 여러 퍼셉트론을 쌓아 만든 신경망으로, 입력층, 은닉층, 출력층으로 구성됩니다.
'''

import numpy as np

# 퍼셉트론 함수 정의
def perceptron(x, w, b):
    """
    x: 입력 값, numpy 배열
    w: 가중치, numpy 배열
    b: 편향, 실수 값
    """
    # 입력 값과 가중치의 곱을 모두 더하고, 편향 값을 더함
    z = np.dot(x, w) + b
    # 활성화 함수 (단위 계단 함수)
    return 1 if z > 0 else 0

# 입력 값
x = np.array([0.5, -0.5])
# 가중치
w = np.array([0.5, 0.5])
# 편향
b = 0.0

# 퍼셉트론 결과 출력
output = perceptron(x, w, b)
print(f'퍼셉트론 출력: {output}')
```

```python
'''
활성화 함수

활성화 함수는 신경망의 출력 값을 비선형으로 변환하여 복잡한 문제를 해결할 수 있도록 합니다.
대표적인 활성화 함수로는 시그모이드(sigmoid), 렐루(ReLU), 탄젠트 하이퍼볼릭(tanh) 등이 있습니다.
'''

def sigmoid(z):
    """
    z: 입력 값, 실수 값
    """
    return 1 / (1 + np.exp(-z))

def relu(z):
    """
    z: 입력 값, 실수 값
    """
    return max(0, z)

def tanh(z):
    """
    z: 입력 값, 실수 값
    """
    return np.tanh(z)

# 예시 입력 값
z = 0.5

# 각 활성화 함수의 결과 출력
print(f'Sigmoid 출력: {sigmoid(z)}')
print(f'ReLU 출력: {relu(z)}')
print(f'Tanh 출력: {tanh(z)}')
```

```python
'''
손실 함수와 최적화 알고리즘

손실 함수는 모델의 예측 값과 실제 값의 차이를 측정하는 함수로, 대표적인 손실 함수로는 평균 제곱 오차(MSE), 교차 엔트로피 손실 등이 있습니다.
최적화 알고리즘은 손실 함수를 최소화하기 위해 가중치를 조정하는 방법으로, 대표적인 알고리즘으로는 경사 하강법(Gradient Descent) 등이 있습니다.
'''

def mean_squared_error(y_true, y_pred):
    """
    y_true: 실제 값, numpy 배열
    y_pred: 예측 값, numpy 배열
    """
    return np.mean((y_true - y_pred) ** 2)

# 예시 실제 값과 예측 값
y_true = np.array([1, 0, 1])
y_pred = np.array([0.8, 0.1, 0.9])

# 평균 제곱 오차 계산
mse = mean_squared_error(y_true, y_pred)
print(f'평균 제곱 오차: {mse}')
```

```python
'''
역전파 알고리즘

역전파 알고리즘은 신경망의 가중치를 학습시키는 방법으로, 출력층에서 입력층으로 오차를 전파하여 가중치를 업데이트합니다.
이는 경사 하강법을 사용하여 손실 함수를 최소화하는 과정입니다.



순전파 (Forward Pass)

선형 결합: z = np.dot(x, w) + b

입력 데이터 x와 가중치 w를 곱하고, 편향 b를 더합니다.
np.dot(x, w)는 입력 데이터와 가중치의 내적을 계산합니다.
z는 활성화 함수에 입력될 선형 결합 값입니다.
활성화 함수 적용: y_pred = sigmoid(z)

z 값을 시그모이드 함수에 통과시켜 y_pred를 계산합니다.
시그모이드 함수는 값을 0과 1 사이로 변환하여 예측 확률을 나타냅니다.
손실 함수의 기울기 계산

손실 함수의 기울기: loss_derivative = y_pred - y_true
예측값 y_pred와 실제 값 y_true의 차이를 계산합니다.
이 값은 손실 함수의 기울기로, 현재 예측이 얼마나 잘못되었는지를 나타냅니다.
역전파 (Backpropagation)

가중치 기울기: w_gradient = loss_derivative * sigmoid_derivative(z) * x

sigmoid_derivative(z)는 시그모이드 함수의 기울기입니다. 이는 z 값에 대한 시그모이드 함수의 민감도를 나타냅니다.
loss_derivative * sigmoid_derivative(z)는 손실 함수의 기울기와 활성화 함수의 기울기를 곱하여 각 가중치가 손실 함수에 미치는 영향을 계산합니다.
이 값을 입력 x와 곱하여 각 가중치에 대한 기울기를 구합니다.
편향 기울기: b_gradient = loss_derivative * sigmoid_derivative(z)

편향의 기울기는 손실 함수의 기울기와 시그모이드 함수의 기울기를 곱하여 계산됩니다.
편향은 모든 입력에 대해 동일하게 적용되므로 입력 x가 필요 없습니다.
가중치와 편향 업데이트

가중치 업데이트: w -= learning_rate * w_gradient

계산된 가중치 기울기를 사용하여 가중치를 업데이트합니다.
learning_rate는 학습 속도를 조절하는 하이퍼파라미터입니다.
편향 업데이트: b -= learning_rate * b_gradient

계산된 편향 기울기를 사용하여 편향을 업데이트합니다.
학습률을 곱하여 업데이트 크기를 조절합니다.
'''

def sigmoid_derivative(z):
    """
    시그모이드 함수의 미분
    z: 입력 값, 실수 값
    """
    return sigmoid(z) * (1 - sigmoid(z))

# 예시 입력 값, 실제 값, 가중치, 편향
x = np.array([0.5, -0.5])
y_true = 1
w = np.array([0.5, 0.5])
b = 0.0

# 순전파 계산
z = np.dot(x, w) + b
y_pred = sigmoid(z)

# 손실 함수의 미분
loss_derivative = y_pred - y_true # 실제로 손실 함수의 기울기를 구하는 방식 중 하나일 뿐 (절대적인것 X)

# 역전파 계산 (가중치와 편향의 변화량 계산)
w_gradient = loss_derivative * sigmoid_derivative(z) * x
b_gradient = loss_derivative * sigmoid_derivative(z)

# 학습률 설정
learning_rate = 0.1

# 가중치와 편향 업데이트
w -= learning_rate * w_gradient
b -= learning_rate * b_gradient

print(f'업데이트된 가중치: {w}')
print(f'업데이트된 편향: {b}')

# 역전파가 어떤식으로 진행되는지 다시 확인
```

```python
'''
TensorFlow와 Keras

TensorFlow는 구글에서 개발한 오픈소스 딥러닝 프레임워크로, 복잡한 수학적 연산과 모델 학습을 효율적으로 수행할 수 있습니다.
 Keras는 TensorFlow 위에서 작동하는 고수준 API로, 신경망 모델을 더 쉽게 만들고 학습시킬 수 있도록 도와줍니다.
'''
#(설치코드)
# pip install tensorflow
# import tensorflow as tf
# print(tf.__version__)


# Keras는 신경망 모델을 간단하게 구성하고 학습시킬 수 있도록 다양한 기능을 제공합니다.

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Sequential 모델을 생성
model = Sequential()

# 첫 번째 은닉층 추가 (Dense 레이어)
model.add(Dense(units=32, activation='relu', input_shape=(784,)))
# 두 번째 은닉층 추가
model.add(Dense(units=32, activation='relu'))
# 출력층 추가
model.add(Dense(units=10, activation='softmax'))

# 모델 요약 출력
model.summary()
```

```python
'''
Keras로 신경망 모델 구성하기

Keras를 사용해 신경망 모델을 구성하는 방법은 매우 간단합니다.
Sequential 모델에 레이어를 차례로 추가하면 됩니다. 각 레이어의 역할을 설명하겠습니다.
'''
# Sequential 모델을 생성합니다.
model = Sequential()

# 첫 번째 은닉층 추가
# units: 뉴런의 수를 의미합니다. 여기서는 32개의 뉴런을 사용합니다.
# activation: 활성화 함수입니다. 'relu'는 Rectified Linear Unit으로, 입력이 0보다 크면 그대로 출력하고, 그렇지 않으면 0을 출력합니다.
# input_shape: 입력 데이터의 형태입니다. (784,)는 784차원의 벡터를 의미합니다.
model.add(Dense(units=32, activation='relu', input_shape=(784,)))

# 두 번째 은닉층 추가
model.add(Dense(units=32, activation='relu'))

# 출력층 추가
# softmax 활성화 함수는 출력값을 확률로 변환하여 다중 클래스 분류 문제에 사용됩니다.
model.add(Dense(units=10, activation='softmax'))

# 모델 컴파일
# optimizer: 최적화 알고리즘입니다. 'adam'은 Adam Optimizer로, 경사 하강법을 기반으로 한 알고리즘입니다.
# loss: 손실 함수입니다. 'sparse_categorical_crossentropy'는 클래스가 정수로 인코딩된 다중 클래스 분류 문제에 사용됩니다.
# metrics: 평가 지표입니다. 여기서는 'accuracy'를 사용하여 모델의 정확도를 평가합니다.
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 모델의 구조를 출력합니다.
model.summary()
```

```python
'''
모델 훈련 및 평가
Keras에서 모델을 훈련하고 평가하는 방법을 알아보겠습니다.
MNIST 데이터셋을 사용하여 예제를 보여드리겠습니다.
'''
from tensorflow.keras.datasets import mnist

# MNIST 데이터셋 로드
# x_train, y_train: 훈련 데이터와 레이블
# x_test, y_test: 테스트 데이터와 레이블
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 데이터 전처리
# x_train과 x_test의 차원을 (샘플 수, 784)로 변환하고, 0-1 범위로 정규화합니다.
x_train = x_train.reshape(-1, 784).astype('float32') / 255
x_test = x_test.reshape(-1, 784).astype('float32') / 255

# 모델 훈련
# fit 함수는 모델을 훈련시키는 역할을 합니다.
# x: 입력 데이터, y: 레이블, epochs: 에포크 수, batch_size: 배치 크기, validation_split: 검증 데이터 비율
# epochs: 전체 데이터셋을 몇 번 반복해서 학습할지를 의미합니다. 에포크 수가 많을수록 모델이 더 많이 학습하지만, 너무 많으면 과적합이 발생할 수 있습니다.
# batch_size: 한 번의 훈련에서 사용할 데이터 샘플의 수를 의미합니다. 배치 크기가 작을수록 메모리 사용량이 적지만 훈련 시간이 길어질 수 있습니다.
# validation_split: 전체 훈련 데이터 중 검증 데이터로 사용할 비율을 의미합니다.
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# 모델 평가
# evaluate 함수는 테스트 데이터에 대한 모델의 성능을 평가합니다.
# x: 테스트 데이터, y: 테스트 레이블
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f'테스트 정확도: {test_accuracy}')
```

```python
'''
과적합 방지 및 모델 개선
과적합을 방지하기 위해 다양한 기법을 사용할 수 있습니다.
대표적으로 드롭아웃(dropout), 조기 종료(early stopping) 등이 있습니다.


드롭아웃(Dropout)
드롭아웃은 무엇인가요?

드롭아웃은 신경망 모델의 학습 과정에서 무작위로 일부 뉴런을 비활성화(또는 "드롭")하는 기법입니다.
이를 통해 모델이 특정 뉴런에 과도하게 의존하지 않도록 하고, 다양한 조합의 뉴런을 학습하도록 유도합니다.

드롭아웃이 어떻게 작동하나요?

학습 과정 중에 드롭아웃 레이어가 무작위로 일부 뉴런을 비활성화합니다.
이 때, 비활성화되는 뉴런의 비율을 rate 매개변수로 설정할 수 있습니다.
예를 들어, 드롭아웃 비율이 0.5이면, 각 학습 단계에서 뉴런의 50%가 무작위로 비활성화됩니다.
이렇게 하면, 모델은 모든 뉴런이 항상 활성화된 상태로 학습하지 않고, 다양한 조합을 통해 학습하게 됩니다.


조기 종료(Early Stopping)
조기 종료는 무엇인가요?

조기 종료는 모델의 학습을 조기에 중단하는 기법입니다.
이는 검증 데이터의 성능이 개선되지 않으면 학습을 멈추고, 과적합을 방지하는 데 도움을 줍니다.
조기 종료가 어떻게 작동하나요?

학습 중에 검증 데이터의 손실(val_loss)을 모니터링합니다.
설정한 patience 값만큼 검증 손실이 개선되지 않으면, 학습을 중단합니다.
예를 들어, patience=3으로 설정하면, 검증 손실이 3번 연속으로 개선되지 않을 때 학습이 조기에 종료됩니다.

'''

from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Sequential 모델을 생성합니다.
model = Sequential()

# 첫 번째 은닉층 추가와 드롭아웃
# Dropout 레이어는 무작위로 일부 뉴런을 학습에서 제외시켜 과적합을 방지합니다.
model.add(Dense(units=32, activation='relu', input_shape=(784,)))
model.add(Dropout(rate=0.5))  # 드롭아웃 비율 50%
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(rate=0.5))

# 출력층 추가
model.add(Dense(units=10, activation='softmax'))

# 모델 컴파일
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 조기 종료 콜백 설정
# EarlyStopping 콜백은 검증 데이터의 손실이 개선되지 않으면 훈련을 조기에 종료합니다.
# monitor: 모니터링할 값, patience: 개선되지 않은 에포크 수
early_stopping = EarlyStopping(monitor='val_loss', patience=3)

# 모델 훈련
model.fit(x_train, y_train, epochs=50, batch_size=32, validation_split=0.2, callbacks=[early_stopping])

# 모델 평가
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f'테스트 정확도: {test_accuracy}')

```

```python
'''
Convolutional Neural Networks (CNN)


컨볼루션 레이어 (Convolutional Layer):

역할: 입력 데이터(예: 이미지)에서 특징을 추출합니다.
작동 원리: 필터(또는 커널)라고 하는 작은 행렬을 사용하여 입력 데이터에 대해 컨볼루션 연산을 수행합니다. 필터는 이미지의 일부분을 스캔하며, 지역적 패턴(예: 엣지, 텍스처 등)을 인식합니다.
예시: 3x3 필터를 사용하여 이미지의 모든 픽셀에 적용하면, 필터는 이미지의 엣지나 패턴을 감지할 수 있습니다.


풀링 레이어 (Pooling Layer):

역할: 특징 맵의 차원을 줄이고, 계산량을 줄이며, 중요한 정보를 유지합니다.
작동 원리: 일반적으로 최대 풀링(max pooling)을 사용합니다. 이는 일정 크기의 영역에서 최대값을 선택하여 출력합니다. 예를 들어, 2x2 최대 풀링은 2x2 영역에서 가장 큰 값을 선택하여 출력합니다.
예시: 2x2 풀링 레이어는 이미지의 크기를 절반으로 줄이면서 가장 중요한 특징을 유지합니다.


완전 연결 레이어 (Fully Connected Layer):

역할: CNN의 최종 단계에서 추출된 특징을 바탕으로 분류 작업을 수행합니다.
작동 원리: 모든 입력 뉴런이 모든 출력 뉴런과 연결됩니다. 이 레이어는 최종적인 분류를 위해 추출된 특징을 통합합니다.

'''
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# CNN 모델 생성
model = Sequential()

# 컨볼루션 레이어 추가: 32개의 3x3 필터를 사용하고 ReLU 활성화 함수 적용
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))

# 최대 풀링 레이어 추가: 2x2 영역에서 최대값을 선택
model.add(MaxPooling2D(pool_size=(2, 2)))

# 두 번째 컨볼루션 레이어 추가: 64개의 3x3 필터를 사용
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))

# 두 번째 최대 풀링 레이어 추가
model.add(MaxPooling2D(pool_size=(2, 2)))

# 특징 맵을 1차원으로 변환
model.add(Flatten())

# 완전 연결 레이어 추가
model.add(Dense(units=128, activation='relu'))

# 출력층 추가: 10개의 클래스에 대한 확률을 출력
model.add(Dense(units=10, activation='softmax'))

# 모델 컴파일
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 모델 구조 출력
model.summary()
```

```python
'''
Recurrent Neural Networks (RNN)

RNN은 시계열 데이터나 순서가 중요한 데이터를 처리하는 데 적합합니다.
RNN은 이전의 상태를 기억하여 현재의 입력과 결합하여 새로운 출력을 생성합니다.
이러한 특성 덕분에 RNN은 자연어 처리, 음성 인식, 시계열 예측 등에서 유용합니다.

RNN의 주요 구성 요소

순환 유닛 (Recurrent Unit):

역할: 시간에 따른 입력 데이터의 순서와 상태를 유지합니다.
작동 원리: 현재의 입력과 이전의 상태를 결합하여 새로운 상태를 생성합니다. 이 상태는 다음 시간 단계로 전달됩니다.
예시: 텍스트 데이터의 경우, 이전 단어의 정보를 기억하여 다음 단어를 예측합니다.


LSTM (Long Short-Term Memory):

역할: RNN의 기본 문제인 장기 의존성 문제를 해결합니다.
작동 원리: LSTM은 장기와 단기 정보를 모두 기억할 수 있는 구조를 가집니다. 셀 상태와 게이트 메커니즘을 통해 정보를 선택적으로 기억하거나 잊어버릴 수 있습니다.


GRU (Gated Recurrent Unit):

역할: LSTM의 간소화된 버전으로, 비슷한 장기 의존성 문제를 해결합니다.
작동 원리: LSTM과 유사하지만 더 간단한 구조로 정보를 처리합니다. GRU는 업데이트 게이트와 리셋 게이트를 사용하여 정보를 조절합니다.

'''

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

# RNN 모델 생성
model = Sequential()

# SimpleRNN 레이어 추가: 50개의 유닛을 사용하고 ReLU 활성화 함수 적용
model.add(SimpleRNN(units=50, activation='relu', input_shape=(28, 28)))

# 완전 연결 레이어 추가
model.add(Dense(units=64, activation='relu'))

# 출력층 추가: 10개의 클래스에 대한 확률을 출력
model.add(Dense(units=10, activation='softmax'))

# 모델 컴파일
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 모델 구조 출력
model.summary()
```

```python
'''
Tensor와 Autograd

Tensor는 다차원 배열을 나타내며, Autograd는 자동으로 미분을 계산해주는 기능입니다.
이는 딥러닝 모델을 학습시키는 데 중요한 역할을 합니다.

Tensor: 데이터를 저장하는 객체입니다. 기본적인 텐서는 CPU에서 작업하지만, GPU로 이동할 수도 있습니다.

Autograd: PyTorch의 자동 미분 기능으로, 역전파(Backpropagation)를 통해 신경망의 가중치를 업데이트합니다.
'''

# Autograd를 이용한 텐서 생성
x = torch.tensor([2.0, 3.0], requires_grad=True)  # requires_grad=True로 설정

# 연산 수행
y = x * 2
z = y.mean()

# 역전파 수행
z.backward()

# 기울기 확인
print(x.grad)  # 출력: tensor([2., 2.])
```

```python
'''
PyTorch로 신경망 모델 구성하기

PyTorch에서 신경망을 구성하기 위해 torch.nn 모듈을 사용합니다.
이 모듈에는 신경망을 만들고 학습시키는 데 필요한 다양한 레이어와 함수가 포함되어 있습니다.

평탄화(flattening)란 다차원 배열을 1차원 배열로 변환하는 과정을 말합니다.
여기서는 28x28 크기의 이미지를 신경망에 입력하기 위해 784 (28 * 28) 차원의 1차원 벡터로 변환합니다.
신경망의 입력층은 1차원 벡터를 받아들이므로, 2차원 이미지 데이터를 1차원으로 변환해주는 과정이 필요합니다.
'''

import torch
import torch.nn as nn
import torch.optim as optim

# 간단한 신경망 모델 정의
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        # nn.Module의 초기화 메소드를 호출하여 부모 클래스의 속성과 메소드를 가져옴
        self.fc1 = nn.Linear(28 * 28, 128)  # 입력: 28x28 이미지(784차원), 출력: 128
        self.fc2 = nn.Linear(128, 10)       # 입력: 128, 출력: 10 (클래스 개수)

    def forward(self, x):
        # x의 크기를 (batch_size, 28*28)로 평탄화
        x = x.view(-1, 28 * 28)
        # x를 2차원으로 만들어줌, 여기서 -1은 batch_size를 의미
        # 예를 들어, (batch_size, 1, 28, 28) -> (batch_size, 784)
        x = torch.relu(self.fc1(x))  # ReLU 활성화 함수 적용
        x = self.fc2(x)              # 출력층
        return x

# 모델 인스턴스 생성
model = SimpleNN()

# 손실 함수와 최적화 알고리즘 정의
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
'실제 이미지 파일 적용해보는 코드 실제로는 많은 반복과 데이터가 필요 할때마다 바뀜'

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from PIL import Image

# 1. 신경망 모델 정의
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(28 * 28, 128)  # 입력: 28x28 이미지, 출력: 128
        self.fc2 = nn.Linear(128, 10)       # 입력: 128, 출력: 10 (클래스 개수)

    def forward(self, x):
        x = x.view(-1, 28 * 28)  # 평탄화: 2D 이미지를 1D 벡터로 변환
        x = torch.relu(self.fc1(x))  # ReLU 활성화 함수 적용
        x = self.fc2(x)              # 출력층
        return x

# 2. 이미지 불러오기
image_path = '이미지1.jpg'  # 실제 이미지 파일 경로
image = Image.open(image_path)

# 3. 이미지 전처리
transform = transforms.Compose([
    transforms.Resize((28, 28)),    # 이미지 크기 조정
    transforms.Grayscale(),         # 흑백 이미지로 변환
    transforms.ToTensor(),          # 텐서로 변환
    transforms.Normalize((0.5,), (0.5,))  # 정규화
])

# 전처리 적용
image_tensor = transform(image)
image_tensor = image_tensor.unsqueeze(0)  # 배치 차원 추가

# 4. 모델 정의 및 학습 (간단한 예제, 실제로는 많은 데이터와 반복이 필요)
model = SimpleNN()
criterion = nn.CrossEntropyLoss()  # 손실 함수
optimizer = optim.Adam(model.parameters(), lr=0.001)  # 옵티마이저

# 모델 학습 (훈련 과정)
for epoch in range(1):  # 1 에포크 (한 번만 학습)
    model.train()  # 학습 모드로 전환
    optimizer.zero_grad()  # 기울기 초기화
    dummy_input = torch.randn(1, 1, 28, 28)  # 랜덤 입력 데이터
    dummy_label = torch.tensor([1])  # 랜덤 라벨
    output = model(dummy_input)  # 모델 예측
    loss = criterion(output, dummy_label)  # 손실 계산
    loss.backward()  # 기울기 계산
    optimizer.step()  # 가중치 업데이트

# 5. 모델 평가
model.eval()  # 평가 모드로 전환
with torch.no_grad():  # 그래디언트 계산 비활성화
    output = model(image_tensor)  # 이미지에 대한 예측 수행
    predicted_class = torch.argmax(output, dim=1).item()  # 가장 높은 확률을 가진 클래스를 예측
    print(f'Predicted class: {predicted_class}')  # 예측된 클래스 출력
```

```python
'''
모델 훈련 및 평가

에포크: 전체 데이터셋을 한 번 훈련하는 과정입니다.
예를 들어, 10 에포크는 전체 데이터셋을 10번 반복하여 훈련하는 것을 의미합니다.

배치 사이즈: 한 번의 훈련 단계에서 처리하는 데이터 샘플의 수입니다.
예를 들어, 배치 사이즈가 32이면, 모델은 32개의 샘플을 한 번에 처리합니다.
'''

# 데이터 로더 정의 (예: MNIST 데이터셋)
from torchvision import datasets, transforms

train_loader = torch.utils.data.DataLoader(
    datasets.MNIST('data', train=True, download=True, transform=transforms.ToTensor()),
    batch_size=64, shuffle=True
)

# 모델 훈련
for epoch in range(10):  # 10 에포크 동안 훈련
    model.train()  # 모델을 훈련 모드로 설정
    for batch_idx, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()  # 기울기 초기화
        output = model(data)  # 모델의 출력
        loss = criterion(output, target)  # 손실 계산
        loss.backward()  # 역전파
        optimizer.step()  # 가중치 업데이트

    print(f'Epoch {epoch+1}, Loss: {loss.item()}')

# 모델 평가
model.eval()  # 평가 모드로 설정
test_loader = torch.utils.data.DataLoader(
    datasets.MNIST('data', train=False, transform=transforms.ToTensor()),
    batch_size=1000, shuffle=False
)

correct = 0
total = 0
with torch.no_grad():  # 기울기 계산 비활성화
    for data, target in test_loader:
        output = model(data)
        _, predicted = torch.max(output.data, 1)
        total += target.size(0)
        correct += (predicted == target).sum().item()

print(f'Accuracy: {100 * correct / total}%')
```

```python
'''
PyTorch

CNN (Convolutional Neural Networks)
'''

import torch.nn as nn

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1)  # 입력 채널 1, 출력 채널 32
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)  # 2x2 최대 풀링
        self.fc1 = nn.Linear(32 * 14 * 14, 128)  # 입력: 32x14x14, 출력: 128
        self.fc2 = nn.Linear(128, 10)  # 10개 클래스

    def forward(self, x):
        x = self.conv1(x)
        x = torch.relu(x)
        x = self.pool(x)
        x = x.view(-1, 32 * 14 * 14)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# 모델 인스턴스 생성
model = CNN()
```

```python
'''
PyTorch

RNN (Recurrent Neural Networks)

RNN은 시퀀스 데이터(예: 텍스트, 음성 등)의 순서를 처리하는 데 유용한 신경망입니다.
RNN은 시간에 따라 데이터를 처리할 수 있는 기능이 있어, 과거의 정보를 기억하고 이를 현재의 예측에 활용할 수 있습니다.

RNN의 기본 원리는 이전 상태의 출력을 다음 상태의 입력으로 사용하는 것입니다.
이 과정에서 순환 구조가 생깁니다.
예를 들어, 텍스트를 처리할 때, 문장의 각 단어를 순서대로 입력하여 문맥을 이해합니다.
'''

import torch.nn as nn
# input_size: 입력 데이터의 특성 수입니다.
# hidden_size: RNN의 숨겨진 상태의 차원입니다. 모델의 복잡성을 결정합니다.
# num_layers: RNN의 레이어 수입니다.
# batch_first=True: 입력 데이터의 첫 번째 차원이 배치 차원임을 의미합니다.
class RNN(nn.Module):
    def __init__(self):
        super(RNN, self).__init__()
        self.rnn = nn.RNN(input_size=10, hidden_size=20, num_layers=1, batch_first=True)
        # input_size: 입력 데이터의 특성 수입니다.
        # hidden_size: RNN의 숨겨진 상태의 차원입니다. 모델의 복잡성을 결정합니다.
        # num_layers: RNN의 레이어 수입니다.
        # batch_first=True: 입력 데이터의 첫 번째 차원이 배치 차원임을 의미합니다.

        self.fc = nn.Linear(20, 1)  # RNN의 출력을 1개의 예측값으로 변환

    def forward(self, x):
        out, _ = self.rnn(x)  # RNN 레이어를 통과
        out = self.fc(out[:, -1, :])  # 마지막 시점의 출력을 가져와서 Fully Connected 레이어에 전달
        return out

# 모델 인스턴스 생성
model = RNN()


```

```python
# 이동평균 필터

import numpy as np
import matplotlib.pyplot as plt

# 예제 데이터 생성
np.random.seed(0)
data = np.sin(np.linspace(0, 20, 100)) + np.random.normal(0, 0.5, 100)  # Sin 곡선에 노이즈 추가

# 이동평균 필터 적용
window_size = 5  # 필터의 윈도우 크기
filtered_data = np.convolve(data, np.ones(window_size)/window_size, mode='valid')
# np.convolve: 두 배열의 합성곱을 계산합니다.
# np.ones(window_size)/window_size: 윈도우 크기만큼의 평균 값을 가지는 필터 생성
# mode='valid': 필터링 후, 원본 데이터와 동일한 길이의 결과를 반환

# 결과 시각화
plt.plot(data, label='Original Data')  # 원본 데이터
plt.plot(range(window_size-1, len(filtered_data)+window_size-1), filtered_data, label='Filtered Data', color='red')
# 필터링된 데이터와 원본 데이터를 시각화
plt.title('Moving Average Filter')
plt.legend()
plt.show()
```

```python
# 버터워스 필터

from scipy.signal import butter, filtfilt
import numpy as np
import matplotlib.pyplot as plt

# 버터워스 필터 디자인
def butterworth_filter(data, cutoff, fs, order=4):
    nyquist = 0.5 * fs  # 나이퀴스트 주파수 계산
    normal_cutoff = cutoff / nyquist  # 정규화된 컷오프 주파수
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    # butter: 버터워스 필터의 계수를 계산합니다.
    # order: 필터의 차수
    # normal_cutoff: 정규화된 컷오프 주파수
    # btype='low': 저주파 필터
    # analog=False: 디지털 필터를 디자인합니다.
    y = filtfilt(b, a, data)  # filtfilt: 필터를 데이터에 적용하여 지연이 없는 필터링 수행
    return y

# 예제 데이터 생성
np.random.seed(0)
data = np.sin(np.linspace(0, 20, 500)) + np.random.normal(0, 0.5, 500)  # Sin 곡선에 노이즈 추가

# 필터링
cutoff_freq = 2.0  # 컷오프 주파수 (Hz)
sampling_rate = 50.0  # 샘플링 주파수 (Hz)
filtered_data = butterworth_filter(data, cutoff_freq, sampling_rate)

# 결과 시각화
plt.plot(data, label='Original Data')  # 원본 데이터
plt.plot(filtered_data, label='Filtered Data', color='red')  # 필터링된 데이터
plt.title('Butterworth Filter')
plt.legend()
plt.show()
```

```python
# 하이패스 필터

from scipy.signal import butter, filtfilt
import numpy as np
import matplotlib.pyplot as plt

# 하이패스 필터 디자인
def highpass_filter(data, cutoff, fs, order=4):
    nyquist = 0.5 * fs  # 나이퀴스트 주파수 계산
    normal_cutoff = cutoff / nyquist  # 정규화된 컷오프 주파수
    b, a = butter(order, normal_cutoff, btype='high', analog=False)
    # butter: 하이패스 필터의 계수를 계산합니다.
    # order: 필터의 차수
    # normal_cutoff: 정규화된 컷오프 주파수
    # btype='high': 고주파 필터
    # analog=False: 디지털 필터를 디자인합니다.
    y = filtfilt(b, a, data)  # filtfilt: 필터를 데이터에 적용하여 지연이 없는 필터링 수행
    return y

# 예제 데이터 생성
np.random.seed(0)
data = np.sin(np.linspace(0, 20, 500)) + np.random.normal(0, 0.5, 500)  # Sin 곡선에 노이즈 추가

# 필터링
cutoff_freq = 2.0  # 컷오프 주파수 (Hz)
sampling_rate = 50.0  # 샘플링 주파수 (Hz)
filtered_data = highpass_filter(data, cutoff_freq, sampling_rate)

# 결과 시각화
plt.plot(data, label='Original Data')  # 원본 데이터
plt.plot(filtered_data, label='Filtered Data', color='red')  # 필터링된 데이터
plt.title('Highpass Filter')
plt.legend()
plt.show()
```

```python
# 보간 및 샘플링

from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt

# 예제 데이터 생성
x = np.linspace(0, 10, 10)  # 기존 x 값 (샘플 데이터의 x 좌표)
y = np.sin(x)  # 기존 y 값 (샘플 데이터의 y 좌표)

# 보간 함수 생성
f_linear = interp1d(x, y, kind='linear')  # 선형 보간 함수
f_cubic = interp1d(x, y, kind='cubic')  # 3차원 보간 함수

# 새로운 x 값 생성 (업샘플링)
x_new = np.linspace(0, 10, 50)  # 더 많은 샘플을 위한 새로운 x 값

# 보간 수행
y_linear = f_linear(x_new)  # 선형 보간을 수행하여 새로운 y 값 생성
y_cubic = f_cubic(x_new)  # 3차원 보간을 수행하여 새로운 y 값 생성

# 결과 시각화
plt.plot(x, y, 'o', label='Original Data')  # 원본 데이터 포인트를 원으로 표시
plt.plot(x_new, y_linear, '-', label='Linear Interpolation')  # 선형 보간된 데이터
plt.plot(x_new, y_cubic, '--', label='Cubic Interpolation')  # 3차원 보간된 데이터
plt.title('Interpolation')
plt.legend()
plt.show()
```

```python
# 층이 여러 개인 신경망 모델

import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
import matplotlib.pyplot as plt

# 1. 데이터 로드 및 전처리
(x_train, y_train), (x_test, y_test) = mnist.load_data()  # MNIST 데이터셋 로드

# 1.1. 데이터 전처리
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0  # 4D 텐서로 변형 및 정규화
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0  # 4D 텐서로 변형 및 정규화
y_train = to_categorical(y_train, 10)  # 레이블을 원-핫 인코딩
y_test = to_categorical(y_test, 10)  # 레이블을 원-핫 인코딩

# 1.2. 데이터 샘플 시각화
plt.imshow(x_train[0].reshape(28, 28), cmap='gray')  # 첫 번째 훈련 데이터 샘플을 흑백 이미지로 시각화
plt.title(f"Label: {y_train[0]}")  # 레이블 출력
plt.show()

# 2. 신경망 모델 정의 및 컴파일
model = Sequential([  # Sequential 모델 생성
    # 첫 번째 2D Convolutional Layer
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),  # 32개의 필터, 필터 크기 3x3, ReLU 활성화
    MaxPooling2D(pool_size=(2, 2)),  # 풀링 크기 2x2
    # 두 번째 2D Convolutional Layer
    Conv2D(64, (3, 3), activation='relu'),  # 64개의 필터, 필터 크기 3x3, ReLU 활성화
    MaxPooling2D(pool_size=(2, 2)),  # 풀링 크기 2x2
    # Flatten Layer
    Flatten(),  # 2D 출력을 1D 벡터로 변환
    # Fully Connected Layer
    Dense(128, activation='relu'),  # 128개의 노드, ReLU 활성화
    # 출력층
    Dense(10, activation='softmax')  # 10개의 클래스, softmax 활성화
])

# 2.1. 모델 컴파일
model.compile(optimizer='adam',  # Adam 옵티마이저 사용
              loss='categorical_crossentropy',  # 다중 클래스 분류를 위한 손실 함수
              metrics=['accuracy'])  # 정확도 지표

# 3. 모델 훈련 및 평가
history = model.fit(x_train, y_train,
                    epochs=5,           # 전체 데이터셋에 대해 5번 반복 학습
                    batch_size=64,      # 한 번에 64개의 샘플로 학습
                    validation_split=0.2)  # 훈련 데이터의 20%를 검증 데이터로 사용

# 3.1. 모델 평가
test_loss, test_accuracy = model.evaluate(x_test, y_test)  # 테스트 데이터로 모델 평가
print(f"Test accuracy: {test_accuracy:.4f}")  # 테스트 정확도 출력

# 3.2. 훈련 과정 시각화
plt.plot(history.history['accuracy'], label='Accuracy')  # 훈련 정확도
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')  # 검증 정확도
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
