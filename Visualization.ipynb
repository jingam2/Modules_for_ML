```python
'''
Matplotlib

Matplotlib는 파이썬에서 가장 널리 사용되는 시각화 라이브러리 중 하나입니다.
간단한 그래프부터 복잡한 시각화까지 다양한 그래프를 그릴 수 있습니다.
'''
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)  # 0에서 10까지 100개의 균등한 값을 생성
y = np.sin(x)  # x 값에 대한 사인 함수 값 생성

# 선 그래프 그리기
plt.plot(x, y, label='Sine Wave')  # x와 y 값을 사용하여 선 그래프를 그리고, 범례에 'Sine Wave'로 표시
plt.xlabel('X-axis')  # x축 레이블 설정
plt.ylabel('Y-axis')  # y축 레이블 설정
plt.title('Line Plot')  # 그래프 제목 설정
plt.legend()  # 범례 표시
plt.show()  # 그래프 출력
```

```python
import matplotlib.pyplot as plt

# 데이터 생성
categories = ['A', 'B', 'C', 'D']  # 카테고리 이름 리스트
values = [10, 20, 15, 25]  # 각 카테고리에 해당하는 값 리스트

# 막대 그래프 그리기
plt.bar(categories, values)  # 카테고리와 값을 사용하여 막대 그래프를 그림
plt.xlabel('Categories')  # x축 레이블 설정
plt.ylabel('Values')  # y축 레이블 설정
plt.title('Bar Plot')  # 그래프 제목 설정
plt.show()  # 그래프 출력
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
data = np.random.randn(1000)  # 정규 분포를 따르는 1000개의 랜덤 값 생성

# 히스토그램 그리기
plt.hist(data, bins=30, edgecolor='black')  # 데이터와 빈의 개수를 지정하여 히스토그램을 그림, 막대 테두리 색상을 검정으로 설정
plt.xlabel('Value')  # x축 레이블 설정
plt.ylabel('Frequency')  # y축 레이블 설정
plt.title('Histogram')  # 그래프 제목 설정
plt.show()  # 그래프 출력
```

```python
'''
Seaborn
Seaborn은 Matplotlib를 기반으로 한 라이브러리로, 더 간단하고 보기 좋은 그래프를 생성할 수 있습니다.
Seaborn 자체로는 상호작용형 기능을 제공하지 않지만, Plotly와 함께 사용할 수 있습니다.
'''

import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드
tips = sns.load_dataset('tips')  # Seaborn에서 제공하는 'tips' 데이터셋을 로드

# 카테고리형 데이터 시각화
sns.barplot(x='day', y='total_bill', data=tips)  # 요일별로 총 청구 금액을 막대 그래프로 시각화
plt.title('Total Bill by Day')  # 그래프 제목 설정
plt.show()  # 그래프 출력
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드
tips = sns.load_dataset('tips')  # Seaborn에서 제공하는 'tips' 데이터셋을 로드

# 수치형 데이터 시각화
sns.scatterplot(x='total_bill', y='tip', data=tips)  # 총 청구 금액과 팁의 관계를 산점도로 시각화
plt.title('Total Bill vs Tip')  # 그래프 제목 설정
plt.show()  # 그래프 출력
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 생성
flights = sns.load_dataset('flights')  # Seaborn에서 제공하는 'flights' 데이터셋을 로드

# 피벗 테이블 생성
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
# index: 행 인덱스로 사용할 열 이름 ('month')
# columns: 열 인덱스로 사용할 열 이름 ('year')
# values: 피벗 테이블의 값으로 사용할 열 이름 ('passengers')

# 히트맵 그리기
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')
# flights_pivot: 히트맵으로 시각화할 데이터
# annot=True: 각 셀에 값을 표시
# fmt='d': 셀 값의 형식을 정수로 설정
# cmap='YlGnBu': 색상 맵을 'YlGnBu'로 설정

plt.title('Heatmap of Flight Passengers')  # 그래프 제목 설정
plt.show()  # 그래프 출력
```

```python
'''
Plotly

Plotly의 상호작용형 그래프는 사용자가 그래프와 직접 상호작용할 수 있는 다양한 기능을 제공합니다.
 예를 들어, 그래프의 특정 부분을 클릭하거나 마우스를 올리면 추가 정보를 표시하거나, 그래프의 축을 확대/축소할 수 있습니다.
 이러한 기능은 데이터 분석과 시각화에서 매우 유용하며, 데이터의 세부 사항을 더 쉽게 탐색할 수 있게 해줍니다.

툴팁(Tooltip): 마우스를 그래프의 데이터 포인트 위에 올리면 해당 데이터의 상세 정보를 표시합니다.

줌(Zoom): 그래프의 특정 부분을 확대하거나 축소할 수 있습니다.

팬(Pan): 그래프를 드래그하여 이동할 수 있습니다.

클릭 이벤트(Click Events): 특정 데이터 포인트를 클릭하면 추가 정보를 표시하거나 다른 동작을 트리거할 수 있습니다.
'''

import plotly.graph_objects as go  # Plotly의 graph_objects 모듈을 가져옵니다.

# 제품명과 판매량 데이터를 정의합니다.
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [150, 200, 300, 250]

# Bar 객체를 생성하여 데이터를 전달합니다.
data = [go.Bar(x=products, y=sales, text=sales, textposition='auto', hoverinfo='x+y')]

# Figure 객체를 생성하고 데이터를 전달합니다.
fig = go.Figure(data=data)

# 그래프 제목과 축 레이블을 설정합니다.
fig.update_layout(title='Product Sales', xaxis_title='Products', yaxis_title='Sales')

# 그래프를 출력합니다.
fig.show()
```

```python
import plotly.graph_objects as go

# 캠페인명과 클릭 수, 전환율 데이터를 정의합니다.
campaigns = ['Campaign A', 'Campaign B', 'Campaign C', 'Campaign D']
clicks = [1000, 1500, 700, 1200]
conversions = [100, 200, 50, 150]

# Bar 객체를 생성하여 데이터를 전달합니다.
data = [go.Bar(x=campaigns, y=clicks, text=conversions, textposition='auto', hoverinfo='x+y+text', name='Clicks')]

# Figure 객체를 생성하고 데이터를 전달합니다.
fig = go.Figure(data=data)

# 그래프 제목과 축 레이블을 설정합니다.
fig.update_layout(title='Campaign Performance', xaxis_title='Campaigns', yaxis_title='Clicks')

# 그래프를 출력합니다.
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# 샘플 데이터 생성
data = {
    'Date': pd.date_range(start='2023-01-01', periods=100),
    'Stock Price': (pd.Series(range(100)) + pd.Series(range(100)).apply(lambda x: x * 0.5)).cumsum()
}
df = pd.DataFrame(data)  # 데이터프레임 생성

# Scatter 객체를 생성하여 데이터를 전달합니다.
scatter = go.Scatter(
    x=df['Date'],  # x축 데이터 (날짜)
    y=df['Stock Price'],  # y축 데이터 (주식 가격)
    mode='lines+markers',  # 선과 마커를 모두 표시
    marker=dict(size=10, color='blue', symbol='circle'),  # 마커 스타일 설정
    text=df['Stock Price'],  # 툴팁에 표시할 텍스트
    hoverinfo='x+y+text'  # 툴팁에 표시할 정보 (x축, y축, 텍스트)
)

# Figure 객체를 생성하고 데이터를 전달합니다.
fig = go.Figure(data=[scatter])

# 그래프 제목과 축 레이블을 설정합니다.
fig.update_layout(
    title='Interactive Stock Price Over Time',  # 그래프 제목
    xaxis_title='Date',  # x축 제목
    yaxis_title='Stock Price',  # y축 제목
    hovermode='closest'  # 툴팁이 가장 가까운 데이터 포인트에 표시되도록 설정
)

# 클릭 이벤트를 처리하는 함수 정의
def display_click_data(trace, points, state):
    print(f"Clicked on: {points.point_inds}, Date: {df.iloc[points.point_inds]['Date'].values}, Price: {df.iloc[points.point_inds]['Stock Price'].values}")

# 클릭 이벤트를 Figure 객체에 추가
fig.data[0].on_click(display_click_data)

# 그래프를 출력합니다.
fig.show()
```
